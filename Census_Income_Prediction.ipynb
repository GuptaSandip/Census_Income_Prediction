### 2.Data Wrangling and Data Manipulation

```python
'''
As we observe insignificant value `?` in 'workplace','occupation', and 'country', So we need
to replace it with mode value of column.
'''
```

### 8. Splitting Data into x and y

```python
x = df[['age','workplace','fnlwgt','education','education_num','marital_status','occupation','relationship',
              'race','Gender','hours_per_week','country','total_asset']]
```

### 9. Splitting x and y into train and test

```python
# splitting x and y into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

### 10. Importing Required Libraries for Model

```python
# importing Required libraries
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
# creating instance of class
logreg = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC()
dt = DecisionTreeClassifier()
```

### 11. Building Model

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    print('classification report')
    print(classification_report(ytest,ypred))
    print('-'*50)
    print('Training accuracy',model.score(xtrain,ytrain))
    print('Testing accuracy',model.score(xtest,ytest))
    print('accuracy score',accuracy_score(ypred,ytest))
```

### Building logistic regression


```python
# building logistic regression
mymodel(logreg)
```

### Hypertunning Max_Depth value
