### Data Wrangling 

```python
'''
As we observe insignificant value `?` in 'workplace','occupation', and 'country', So we need 
to replace it with mode value of column.
'''
```

```python
# Replacing `?` to model values for 'workplace','occupation', and 'country' columns.
column_to_manipulate = ['workplace','occupation','country']
for col in column_to_manipulate:
    df[col].replace(' ?',df[col].mode()[0],inplace=True)
```

### Features Description

```python
list_of_columns = list(df.columns)
print(f'List of columns:',list_of_columns)
```

- Age: The age of the individual.

- Workplace: The type of workplace or employment status (e.g., private sector, government, self-employed).

- Fnlwgt: Final weight. This is a weighting factor used in survey sampling to adjust the sample to be more representative of the population it's drawn from.

- Education: The highest level of education attained by the individual.

- Education_num: The numerical representation of education level, likely corresponding to the number of years of education completed.

- Marital_status: The marital status of the individual (e.g., married, single, divorced).

- Occupation: The type of occupation or job the individual is engaged in.

- Relationship: The individual's relationship status (e.g., husband, wife, own-child, unmarried).

- Race: The racial or ethnic group the individual identifies with.

- Gender: The gender of the individual.

- Capital_gain: Capital gains recorded for the individual (i.e., profits from the sale of assets like stocks, real estate).

- Capital_loss: Capital losses recorded for the individual (i.e., losses from the sale of assets like stocks, real estate).

- Hours_per_week: The number of hours worked per week.

- Country: The country of origin or current residence of the individual.

- Salary: The salary or income level of the individual.

### Splitting Data into x and y

```python
x = df[['age','workplace','fnlwgt','education','education_num','marital_status','occupation','relationship',
              'race','Gender','hours_per_week','country','total_asset']]
```

```python
y = df.salary
```

### Splitting x and y into train and test

```python
# splitting x and y into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

### Importing Required Libraries for Model

```python
# importing Required libraries
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
# creating instance of class
logreg = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC()
dt = DecisionTreeClassifier()
```

### Building Model

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    print('classification report')
    print(classification_report(ytest,ypred))
    print('-'*50)
    print('Training accuracy',model.score(xtrain,ytrain))
    print('Testing accuracy',model.score(xtest,ytest))
```

### Building logistic regression


```python
# building logistic regression
mymodel(logreg)
```

### Building Logistic regression with hyperparameter

```python
logreg_model_2 = LogisticRegression(solver='liblinear')
```

```python
mymodel(logreg_model_2)
```

### Building knn


```python
# building knn
mymodel(knn)
```

### Building svm


```python
# building svm
mymodel(svm)
```

### Building Descision Tree


```python
# building dt
mymodel(dt)
```

### Hypertunning Max_Depth value

```python
for i in range(1,50):
    dt2 = DecisionTreeClassifier(max_depth=i)
    dt2.fit(xtrain,ytrain)
    ypred=dt2.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    print(f'Max depth value {i}, Accuracy:{ac}')
#     print(pd.DataFrame({'Max depth value':i,'accuracy':ac}))
```

```python
dt3 = DecisionTreeClassifier(max_depth=7)
mymodel(dt3)
```

### Hypertunning Max_Sample_Split value

```python
for i in range(2,50):
    dt3 = DecisionTreeClassifier(min_samples_split=i)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    print(f'min_samples_split {i}, Accuracy:{ac}')
```
